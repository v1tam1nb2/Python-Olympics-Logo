```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots()
radius = 0.5
thickness = 6.0

def draw_ring(color, x, y):


    return plt.Circle((x, y), radius, fill=False, lw=thickness, color=color)


blue_ring = draw_ring("blue", -1.0, 0.4)
black_ring = draw_ring("black", 0.2, 0.4)
red_ring = draw_ring("red", 1.4, 0.4)
yellow_ring = draw_ring("orange", -0.4, -0.1)
green_ring = draw_ring("green", 0.8, -0.1)

# x軸,y軸の表示範囲
plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 1.5)

axes.set_aspect(1)
axes.add_artist(blue_ring)
axes.add_artist(black_ring)
axes.add_artist(red_ring)
axes.add_artist(yellow_ring)
axes.add_artist(green_ring)

plt.title("Olympics Logo")
figure.savefig("logo.png")
```
